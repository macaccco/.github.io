```python
import random

France = [85,86,83]
England = [85,83,82]
Austria = [79,78,77]
Hungary = [74,69,73]

ManCity = France
Chelsea = Austria

p1diff = ManCity[0] - Chelsea[2]
middiff = ManCity[1] - Chelsea[1]
p2diff = ManCity[2] - Chelsea[0]



Attempt = 0
TotalChelsea = 0
TotalCity = 0
Flag = True
while Attempt<1:
    turn = 0
    ChelseaScore = 0
    ManCityScore = 0
    position = "Mid"
    positions = []
    Attempt += 1
    while turn < 20:
        turn += 1
        positions.append(position)
        #print(position, ChelseaScore, ManCityScore)
        if position == "Mid":
            value = random.randint(1,120)
            #print(value)
            if value > (60 + middiff):
                position = "CityMid"
            else:
                position = "ChelseaMid"
        elif position == "CityMid":
            value = random.randint(1,120)
            #print(value)
            if value > (60 + (middiff + p2diff)/2):
                position = "City"
            else:
                position = "Mid"
        elif position == "ChelseaMid":
            value = random.randint(1,120)
            #print(value)
            if value > (60 + (middiff + p1diff)/2):
                position = "Mid"
            else:
                position = "Chelsea"
        elif position == "City":
            value = random.randint(1,120)
            #print(value)
            if value > (60 + p2diff):
                position = "Mid"
                ChelseaScore += 1
            else:
                position = "CityMid"
        elif position == "Chelsea":
            value = random.randint(1,120)
            #print(value)
            if value > (60 + p1diff):
                position = "ChelseaMid"
            else:
                position = "Mid"
                ManCityScore += 1

    print("England: " + str(ChelseaScore) + " France: " + str(ManCityScore))
    TotalChelsea += ChelseaScore
    TotalCity += ManCityScore
    #print(positions.count("City"),positions.count("CityMid"),positions.count("Mid"),positions.count("ChelseaMid"),positions.count("Chelsea"))
    if ChelseaScore>ManCityScore:
        Flag = False
        #print(Attempt)
print("England: " + str(TotalChelsea) + " France: " + str(TotalCity))
print(TotalChelsea/20, TotalCity/20)
```
